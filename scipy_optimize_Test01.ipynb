##**2. Minimization problem with equality and inequality constraints.**

image.png

1. 2nd constraint is wrong in the above snapshot. It is corrcted in the code.
2. Ensure the inequality constraint is "upper" type (>=)

```python
import numpy as np
from scipy.optimize import minimize
```

```python
objective_func = lambda x: x[0]**2+(x[0]*x[1])
constraint1 = {'type':'eq', 'fun': lambda x: x[0]**3+(x[0]*x[1])-100} 
constraint2 = {'type':'ineq', 'fun': lambda x: x[0]**2+x[1]-50} 
#constraint2 = {'type':'ineq', 'fun': lambda x: 50-(x[0]**2+x[1])} 
constraints = [constraint1, constraint2]

x0=[1,1]
bounds = [(-100,100), (-100,100)]

result = minimize(objective_func, x0, method='SLSQP', bounds=bounds, constraints=constraints)
#result = minimize(objective_func, x0, method='L-BFGS-B', bounds=bounds, constraints=constraints)

print(result)
```
