**Curve fitting using Least Squares Minimization**

image.png

```python
import numpy as np

phi = np.array([[0,1], [1,1], [2,1], [3,1], [4,1], [5,1]])
Y= np.array([[15], [10], [9], [6], [2], [0]])

#theta = np.linalg.lstsq(phi, Y)
theta = np.linalg.lstsq(phi, Y)[0]
print(theta)

theta= np.linalg.inv(phi.transpose()*np.mat(phi))*phi.transpose() *Y

print(theta)
```
